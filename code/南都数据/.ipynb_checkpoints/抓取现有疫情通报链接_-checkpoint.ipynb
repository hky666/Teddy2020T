```python
def get(city_url_name):
    c =  []
    url = 'https://m.mp.oeeee.com/h5/pages/v20/nCovcase/{}.html'.format(city_url_name)
    driver.get(url)
    sleep(2)
    r = driver.page_source
    soup = BeautifulSoup(r)
    res =  []

    city = soup.find_all(class_="city-box")
    for ci in city:
        #找到城市名字
        cname = ci.find(class_="header").text.strip()
        qu = ci.find_all(class_="city-card")
        for qui in qu:
            qui_name = qui.find(class_="header").text.strip()
            # 具体地点
    #         print(qui)
            locationL = qui.find_all(class_="li")
            for li in locationL:
                location_name = li.text.strip()
                try:
                    url = li['href']
                except Exception as e:
                    url = ''
                res.append([cname,qui_name,location_name,url])
    return res
```
