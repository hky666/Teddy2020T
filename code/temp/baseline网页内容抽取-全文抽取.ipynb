```python
import requests
from bs4 import BeautifulSoup
import re
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
mpl.rcParams["font.sans-serif"]=["SimHei"]
mpl.rcParams["axes.unicode_minus"]=False
from selenium import webdriver
from  time import sleep
import pandas as pd
import json
from  random import random
import re
```

```python
# data read


df=pd.read_excel('./websites.xlsx')
height,width = df.shape
print(height,width,type(df))
df.rename(columns={'疫情通报网址':'url'}, inplace=True)

df['url'] = df['url'].apply(lambda x: x.replace(' ',''))
```

```python
def get_full_text(url,driver,use_webdriver=False):
    '''基线：网页全部中文的提取'''
    if 'weibo' in url:
        use_webdriver=True

        
    if use_webdriver:
        # 用引擎
        driver.get(url)
        sleep(0.2)
        html =  driver.page_source
    else:
        # 用request
        header = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4121.0 Safari/537.36 Edg/84.0.495.0'
        }
        r = requests.get(url, headers=header)
        r.encoding= r.apparent_encoding
        html = r.text
        
    # 去html的css和脚本
    soup = BeautifulSoup(html, features="html.parser")
    if soup.title is not None:
        title = soup.title.text
    else:
        title = 'unknow'
        
    for script in soup(["script", "style"]):
        script.decompose()
    text = soup.get_text()
    
    return title,text
```

```python
def baseline_extract(url_list,driver):
    tietle_list = []
    text_list = []
    for url in url_list:
        print(url)
        t1,t2 = get_full_text(url,driver)
        tietle_list.append(t1)
        text_list.append(t2)
    data={'url':url_list,'title':tietle_list,'text':text_list}
    df = pd.DataFrame(data)
    return df
```

```python
driver = webdriver.Chrome()
df_text_result = baseline_extract(df['url'],driver)
driver.close()
```

```python
df_text_result.to_csv('./baseline_extract.csv')
```
