```python
import requests
from bs4 import BeautifulSoup
import re
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
mpl.rcParams["font.sans-serif"]=["SimHei"]
mpl.rcParams["axes.unicode_minus"]=False
from selenium import webdriver
from  time import sleep
import pandas as pd
import json
from  random import random
import re
```

```python
driver = webdriver.Chrome()
driver.get('https://weibo.com')

# driver.implicitly_wait(5) # 操作、获取元素时的隐式等待时间
driver.set_page_load_timeout(10) # 页面加载超时等待时间
```

```python
def get_data(url,driver,print_plot=False,print_textdict=False):
    ''' 
        input:url,driver
    
        output:未处理前的html文本,经过html清洗之后的文本,每行文本的字典,每行文本长度的列表
    '''
    html =  ''
    # 用引擎
    driver.get(url)
    sleep(0.2)
    html =  driver.page_source
    
    soup = BeautifulSoup(html, features="html.parser")
    
    for script in soup(["script", "style"]):
        script.decompose()
    text = soup.get_text()
    
    # 每行的字符数目统计
    count = []
    flag = 0
    k = 0
    row = 0
    s = []
    s_c = ''

    for si in text:
        if si != '\n':
            flag = 0 
            k+=1
            s_c += si
        else:
            flag = 1
            count.append(k)
            s.append(s_c)
            k=0
            s_c = ''
            row+=1
    
    if s_c != '':
        flag = 1
        count.append(k)
        s.append(s_c)
        k=0
        s_c = ''
        row+=1

    x = np.arange(1,len(count)+1,1)
    y = np.array(count)
    if print_plot:
        fig = plt.figure()
        ax1 = fig.add_subplot(111)
        #设置标题
        ax1.set_title('文本密度分布函数')
        plt.xlabel("字符行序号")
        plt.ylabel("字符数目")
        #画散点图
        ax1.scatter(x,y,c = 'r',marker = 'o')
        # 限制刻度
        my_x_ticks = np.arange(1, len(count)+1, 1)
        plt.xticks(my_x_ticks)   #设置坐标轴刻度
        #显示所画的图
        plt.show()
    if print_textdict:
        print (s)
    return html,text,s,y
```
