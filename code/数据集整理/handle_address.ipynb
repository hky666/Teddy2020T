```python
items = [] 
import jsonlines
with open("test_submit.json", "r+", encoding="utf8") as f:
    for item in jsonlines.Reader(f):
        items.append(item)
        print(item)
        pass
```

```python
items[:100]
```

```python
addrees = []

for item in items:
    label = item['label']
    if 'address' in label:
        addrees+=(list(label['address'].keys()))
```

```python
addrees
```

```python
astring =  ''
for ai in addrees:
    astring+=ai+'\n'
f=open('location.txt','w+',encoding='utf-8')
f.write(astring)
f.close
```
